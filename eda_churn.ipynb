# Transformação de Dados

```python
# Transformar coluna TotalChages de String para float - Abordagem 1 com astype

df_contract.TotalCharges = df_contract.TotalCharges.astype('float64')
```

```python
# Transformar coluna TotalChages de String para float - Abordagem 2 com to_numeric
df_contract.TotalCharges = pd.to_numeric(df_contract.TotalCharges)
```

```python
# Transformar coluna "TotalCharges" de String para Float - Abordagem 3 (To_NUMERIC + ERRORS='COERCE')
df_contract['TotalCharges'] = pd.to_numeric(df_contract['TotalCharges'], errors='coerce')
```

## Unificar DataFrame De Customers, Services e Contracts

```python
# Contar Quantidade de Registros por DataFrame
print(f'Quantidade de Registros no DataFrame Customers: {len(df_customers)}')
print(f'Quantidade de Registros no DataFrame Services: {len(df_services)}')
print(f'Quantidade de Registros no DataFrame Contracts: {len(df_contract)}')
```

```python
#  Renoemar colunas de todos os DataFrames para o padrão inglês
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
# Validar se a alteração foi aplicada
df_services.info()
```

```python
# Unificar DataFrame de Customers com Services, criando um terceito DataFrame
df_temp=df_customers.merge(df_services, on=['IDCliente'])
```

```python
# Mostrar estrutura do DataFrame temp
df_temp.info()
```

```python
# Mostrar as 5 primeiras linhas do DataFrame temp
df_temp.head(5)
```

```python
# Unificar df_temp com Contracts, usando colunas de junção com nome diferente
df_churn = df_temp.merge(df_contract, left_on=['IDCliente'], right_on=['customerID'])
```

```python
# Validade estrutura do DataFrame final
df_churn.info()
```

```python
# Fazendo um dataFrame final, unficando as 3 tabelas em um unico DataFrame com nomes diferentes
df_churn = df_customers.merge(df_services, on=['IDCliente']).merge(df_contract,
left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
# Removendo coluna de um DataFrame
df_churn.drop(['customerID'], axis=1, inplace=True)
```
